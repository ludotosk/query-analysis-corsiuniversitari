mariadb -hserver -uroot -ppassword --port 6526 --protocol=TCP railway -e "SELECT * FROM clicks;" > clicks.tsv

```python
import pandas as pd
```

```python
ds = pd.read_table('clicks.tsv')
ds.tail(50)
```

```python
ds.time = pd.to_datetime(ds.time)
```

```python
ds[(ds["page"] == 'tutti-i-corsi') | (ds["page"] == 'corsi-di-laurea-in-inglese') | (ds["page"] == 'corsi-senza-test')].groupby(['city'])['city'].count().reset_index(name='count').sort_values(['count'], ascending=False).head(30)
```

```python
ds[ds.time > pd.to_datetime("2023-09-13 14:19:29")].groupby(['class'])['class'].count().reset_index(name='count').sort_values(['count'], ascending=False).head(30)
```
